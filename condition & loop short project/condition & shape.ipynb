```python
num_rows=int(input("enter the number of rows"))
for index in range(0,num_rows):
    for j in range(0,num_rows-index,-1):
        print(end=' ')
    for j in range(0,index+1):
        print("*",end=' ')
    print()
    
```

```python
num_rows=int(input("enter the number of rows"))
for index in range(0,num_rows):
    for j in range(0,num_rows+index,-1):
        print(end=' ')
    for j in range(0,index-1):
        print("*",end=' ')
    print()
    
```

# shape

```python
n=5
for i in range(n):   #for row
    for j in range(n):  #for column
        print('*',end=' ')
    print()
```

```python
n=5
for i in range(n):
    for j in range(i+1):      #increasing
        print('*',end=' ')
    print()
```

```python
n=5
for i in range(n):               #decreasing
    for j in range(i,n):
        print('*',end=' ')
    print()
```

```python
n=5
for i in range(n):
    for j in range(i,n):
        print(' ',end=' ')
    for k in range(i+1):
        print('*',end=' ')
    print()
```

```python
n=5
for i in range(n): 
    for j in range(i,n): 
        print(' ',end=" ")
    for k in range(i+1):
        print('*',end=" ")
    for l in range(i):
        print('*',end=" ") 
    print()
```
