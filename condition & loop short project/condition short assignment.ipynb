# check even odd

```python
value=int(input("enter the number"))
if value%2==0:
    print("even")
else:
    print("odd")
```

# check number positive,negative and zero

```python
value=int(input("enter the number"))
if value>0:
    print("positive")
elif value<0:
    print("negative")
else:
    print("zero")
```

# check character is vowel or consonant

```python
character=str(input("enter a character"))
if character in('a','e','i','o','u','A','E','I','O','U'):
    print("vowel")
else:
    print("consonant")

```

# check leap year or not

```python
year=int(input("enter the year"))
if year%400==0 or year%4==0:
    print("leap year")
else:
    print("not leap year")
```

# grade evaluator

```python
score=int(input("enter your score"))
if score<40:
    print("fail")
elif score>=40 and score<50:
    print("D")
elif score>=50 and score<60:
    print("C")
elif score>=60 and score<70:
    print("B")
elif score>=70 and score<80:
    print("A")
elif score>=80 and score<=100:
    print("A+")
else:
    print("invalid number")
```

# check divisibility

```python
value=int(input("enter your value"))
if value%5==0 and value%11==0:
    print("divisible by 5 and 11")
else:
    print("not divisible by 5 and 11")
```

# check large value

```python
value_1=int(input("enter your first value"))
value_2=int(input("enter your second value"))
value_3=int(input("enter your third value"))
if value_1>value_2 and value_1>value_3:
    print("large first value")
elif value_2>value_1 and value_2>value_3:
    print("large second value")
elif value_3>value_1 and value_3>value_2:
    print("large third value")
else:
    print("hola")
    
```

# traffic light

```python
color=str(input("enter signal color"))
if color=="red" or color=="Red" or color=="RED":
    print("Stop")
elif color=="yellow" or color=="Yellow" or color=="YELLOW":
    print("Ready")
elif color=="green" or color=="Green" or color=="GREEN":
    print("Go")
else:
    print("Invalid color")
```
