```python
# check taka range using and
taka = 99
if 100 <= taka <= 200:
    print("i can buy bag")
elif 200 <= taka <= 300:
    print("i can buy shoes")
else:
    print("you can keep your taka")
```

```python
# take mark input and give grade
mark = int(input("enter your mark: "))
if mark < 40:
    print("fail")
elif 40 <= mark <= 50:
    print("B+")
elif 51 <= mark <= 59:
    print("A-")
elif 60 <= mark <= 69:
    print("A")
elif 70 <= mark <= 79:
    print("A+")
elif 80 <= mark <= 100:
    print("You are outstanding")
else:
    print("your mark is invalid")
```
